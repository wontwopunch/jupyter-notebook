```python
import seaborn as sns
import matplotlib.pyplot as plt

# Seaborn 샘플 데이터셋 불러오기
tips = sns.load_dataset('tips')

# 데이터셋 살펴보기
print(tips.head())

# 기본 스타일 설정 (optional)
sns.set(style='whitegrid')

# 히스토그램 그리기
plt.figure(figsize=(8, 4))
sns.histplot(tips['total_bill'], bins=30, kde=True, color='blue') # kde=True: 커널 밀도 추정 곡선을 함께 그립
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()

# 산점도 그리기
plt.figure(figsize=(8, 4))
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', palette='Set1', s=100) # s=100: 점의 크기, palette='Set1': 색상 팔레트
plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
import warnings
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 예제 데이터 생성 (정규분포)
# loc: 평균, scale: 표준편차
data = np.random.normal(loc=0, scale=1, size=1000)

# 경고 메시지 무시
warnings.filterwarnings("ignore", category=UserWarning, module='seaborn')

# displot 사용 예시
sns.displot(data, bins=30, kde=True)
plt.title("Distribution Plot using displot")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# kdeplot 사용 예시
# kdeplot은 순수하게 KDE만을 시각화하는 데 사용
sns.kdeplot(data, shade=True) # shade=True: KDE 아래 영역이 음영 처리
plt.title("Kernel Density Estimate using kdeplot")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
# 여러 분포도를 한 그래프에 시각화
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

sns.histplot(data, bins=30, kde=True, ax=ax[0])
ax[0].set_title("Histogram with KDE")

sns.kdeplot(data, shade=True, ax=ax[1])
ax[1].set_title("Kernel Density Estimate")

plt.tight_layout()
plt.show()
```

```python
# 카운트플롯(countplot)
# 카테고리 데이터의 빈도를 시각화하는 데 매우 유용

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 예제 데이터 생성
data = {
    'Category': ['A', 'B', 'C', 'A', 'B', 'A', 'C', 'C', 'B', 'A']
}
df = pd.DataFrame(data)

# 카운트 플롯 생성
sns.countplot(x='Category', data=df) # x축에 카테고리 데이터를 설정
plt.title("Count Plot of Categories")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
# 색상 설정
sns.countplot(x='Category', data=df, palette='viridis')
plt.title("Count Plot with Custom Colors")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
# 수직 플롯 생성
sns.countplot(y='Category', data=df, palette='coolwarm')
plt.title("Vertical Count Plot")
plt.xlabel("Count")
plt.ylabel("Category")
plt.show()
```

```python
# hue 옵션 사용
# 예제 데이터 생성
data = {
    'Category': ['A', 'B', 'C', 'A', 'B', 'A', 'C', 'C', 'B', 'A'],
    'Subcategory': ['X', 'Y', 'X', 'Y', 'X', 'Y', 'Y', 'X', 'Y', 'X']
}
df = pd.DataFrame(data)

# hue 옵션 사용
sns.countplot(x='Category', hue='Subcategory', data=df, palette='Set2')
plt.title("Count Plot with Hue")
plt.xlabel("Category")
plt.ylabel("Count")
plt.legend(title='Subcategory')
plt.show()
```

```python
# 페어플롯(pairplot)
# 여러 변수 간의 관계를 시각화하는 데 매우 유용

# 예제 데이터 생성 (Iris 데이터셋 사용)
df = sns.load_dataset('iris')

# 페어 플롯 생성
sns.pairplot(df)
plt.show()
```

```python
# hue 옵션 사용
sns.pairplot(df, hue='species')
plt.show()
```

```python
# 다이어고널에 KDE 추가
sns.pairplot(df, hue='species', diag_kind='kde')
plt.show()
```

```python
# 마커 스타일 변경
sns.pairplot(df, hue='species', markers=['o', 's', 'D'])
plt.show()
```

```python
# 히트맵(heatmap)
# 

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 예제 데이터 생성
np.random.seed(0)
data = pd.DataFrame(np.random.randn(10, 10), columns=list('ABCDEFGHIJ'))

# 히트맵 생성
# data.corr(): 데이터프레임의 각 변수들 간의 상관계수 행렬을 계산
# annot=True: 각 셀에 숫자를 표시합니다 (상관계수 값)
# cmap='coolwarm': 색상 맵을 'coolwarm'으로 설정. 'coolwarm'은 양극단의 상관관계를 분명히 보여줌
# vmin=-1, vmax=1: 색상 맵의 범위를 -1에서 1 사이로 설정 (상관계수의 범위)
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Heatmap of Correlation Matrix")
plt.show()
```

```python
# 셀 사이즈 조정
# square=True: 각 셀을 정사각형으로 설정
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1, square=True)
plt.title("Heatmap of Correlation Matrix with Squared Cells")
plt.show()
```

```python
# 마스크 추가
# mask: 원하는 셀을 숨기는 데 사용. 여기서는 상삼각 행렬(upper triangle)을 마스크로 사용하여 반대편의 데이터를 숨김
mask = np.triu(np.ones_like(data.corr(), dtype=bool))
sns.heatmap(data.corr(), mask=mask, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Heatmap of Correlation Matrix with Mask")
plt.show()
```

```python
# 색상 막대 위치 변경
# cbar_kws={'orientation': 'horizontal'}: 색상 막대의 방향을 가로로 변경
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1, cbar_kws={'orientation': 'horizontal'})
plt.title("Heatmap of Correlation Matrix with Horizontal Colorbar")
plt.show()
```
