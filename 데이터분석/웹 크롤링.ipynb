```python
browser = webdriver.Chrome()
```

```python
url = 'https://www.netflix.com/kr/browse/genre/839338'
browser.get(url)
```

```python
html = browser.page_source
```

```python
! pip install bs4
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
section_list = soup.select('section')
```

```python
len(section_list)
```

```python
section_list
```

```python
# tag.text
section_list[2].text
```

```python
section_list[-5].text
```

```python
section = section_list[3]
section_title = section.select('h2 a span.nm-collections-row-name')
len(section_title)
```

```python
section_title[0].text
```

```python
for section in section_list:
    selected_elements = section.select('h2 a span.nm-collections-row-name')
    if selected_elements:
        section_title = selected_elements[0].text
        print(section_title)
```

```python
# 프로그램 파트
program_list = soup.select('li a span.nm-collections-title-name')
len(program_list)
```

```python
program_list[0].text
```

```python
for program in program_list:
    print(program.text)
```
